```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
import pandas as pd
import time
from lxml import html
from time import sleep

url ='https://www.amazon.eg/s?k=power+bank+anker&language=en&crid=M9WF2AGVEG0&qid=1732723707&sprefix=power+bank+%2Caps%2C716&ref=sr_pg_1'
path = 'Users\HP\Desktop\scrap'
options= webdriver.EdgeOptions()
driver = webdriver.Edge(options=options)
driver.get(url)
page = 'https://www.amazon.eg/s?k=power+bank+anker&language=en&crid=M9WF2AGVEG0&qid=1732723707&sprefix=power+bank+%2Caps%2C716&ref=sr_pg'

product= []
price=[]
rate=[]
company=[]
discount=[]
for i in range (1, 5):
    driver.get(f'{page}_{i}')
    tree = html.fromstring(driver.page_source)
    items = tree.xpath('//div[@data-component-type="s-search-result"]')
    sleep(3)
    for item in items:
        product.append(item.xpath(".//span[@class='a-size-base-plus a-color-base a-text-normal']/text()"))
        try:
            price.append(item.xpath(".//span[@class='a-price-whole']/text()"))
        except:
            price.append('null')
        try:
             rate.append(item.xpath(".//span[@class='a-size-base s-underline-text']/text()"))  
        except:
            rate.append("null")
        try:
             company.append(item.xpath('.//span[@class="a-size-base-plus a-color-base"]/text()'))
        except:
            company.append("null")
        try:
            discount.append(item.xpath('.//div[contains(@class, "a-row a-size-base a-color-secondary")]/span/text()'))
        except:
        
            discount.append("null")  

```

```python
import pandas as pd
df = pd.DataFrame({'Company': company , 'Product':product, 'Price':price , 'Rating':rate , 'Discount':discount})
```
